# Etica, vieses e limitacoes

## Pontos de atencao
- Dados incompletos podem penalizar instituicoes menores ou novas.
- Vies de cobertura: noticias negativas sao mais abundantes que positivas.
- Indicadores de seguranca (CVE, incidentes) nao refletem a maturidade real do controle.

## Mitigacoes recomendadas
- Auditoria humana para decisoes regulatorias ou de alto impacto.
- Revisao periodica de features para remover proxies indevidos.
- Avaliar performance por grupos e ajustar limiares.

## Uso indevido
Este modelo nao deve ser usado isoladamente para decisoes regulatorias.
A recomendacao e usar o Trust Score como sinal auxiliar, com revisao humana.
