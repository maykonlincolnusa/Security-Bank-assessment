```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from models.synthetic import generate_synthetic_dataset

df = generate_synthetic_dataset(rows=800)
df.head()
```

```python
num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
corr = df[num_cols].corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlacoes Numericas')
plt.show()
```

```python
daily = df.groupby('ref_date', as_index=False).agg({'trust_score':'mean','npl_ratio':'mean','security_incidents':'mean'})
daily['ref_date'] = pd.to_datetime(daily['ref_date'])
daily = daily.sort_values('ref_date')
daily.plot(x='ref_date', y=['trust_score','npl_ratio','security_incidents'], figsize=(12,4), title='Series Temporais')
```
